```python
import threading
import socket
import sys
import math

if __name__ == "__main__":
    
    idSocket = socket.socket()
    host = "127.0.0.1"
    port = 12345
    print("Socket creado")

    idSocket.connect((host, port))
    print("Socket conectado al servidor")
    
    dir = "IN/"
    filename = "foto.jpg"
    in_file = open(dir+filename, "rb") # opening for [r]eading as [b]inary
    data = in_file.read() 
    in_file.close()
    
    if int(len(data)/256)<(len(data)/256):
        packages = math.floor(len(data)/256)+1
    else:
        packages = math.floor(len(data)/256)
    
    while len(filename)<256:
        filename = filename + '|'
    packages2 = str(packages)
    while len(packages2)<16:
        packages2 = '0'+packages2 
        
    try:
        idSocket.send(filename.encode())
    except:
        print("Fallo al enviar nombre del archivo")
        sys.exit();
    print("Nombre del archivo enviado")
    
    try:
        idSocket.send(packages2.encode())
    except:
        print("Fallo al enviar datos del archivo")
        sys.exit();
    print("Informaci√≥n sobre el archivo enviada")
    
    for i in range(packages):
        buffer = data[i*256:min(len(data),(i+1)*256)]
        try:
            idSocket.send(buffer)
        except:
            print("Fallo al enviar un paquete de datos: %s/%s"%(i+1, packages))
            break;
        print("%s/%s"%(i+1, packages))
    else:
        print("Archivo enviado correctamente")


    idSocket.close()
    print("Conexi√≥n finalizada")
```
